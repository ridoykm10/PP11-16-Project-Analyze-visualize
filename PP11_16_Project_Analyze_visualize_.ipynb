```python
# prompt: Create a pie chart of the study abroad distribution by country

import pandas as pd
import matplotlib.pyplot as plt

# Assuming your dataframe is named 'df' and already loaded
# If not, load it first using:
# df = pd.read_csv("student_data.csv")

# Count the occurrences of each study abroad country
study_abroad_counts = df['study_abroad_country'].value_counts()

# Create the pie chart
plt.figure(figsize=(8, 8))  # Adjust figure size as needed
plt.pie(study_abroad_counts, labels=study_abroad_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Study Abroad Distribution by Country')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
